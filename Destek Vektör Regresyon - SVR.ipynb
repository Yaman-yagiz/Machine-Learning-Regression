#### SVR algoritması aykırı değerlere karşı dirençsiz olduğundan model oluşturmadan verileri scale etmemiz gerekiyor.
#### SVR daha çok sınıflandırmada kullanılsa da regresyon içinde kullanılabilir.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("maaslar.csv")
# dataframe
x=data.iloc[:,1:2]
y=data.iloc[:,2:]
# numpy array
X=x.values
Y=y.values
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_scaled=sc.fit_transform(X)
y_scaled=np.ravel(sc.fit_transform(Y.reshape(-1,1)))
```

```python
from sklearn.svm import SVR
svr=SVR(kernel='rbf')
svr.fit(x_scaled,y_scaled)
print(svr.predict([[6.6]]))
```

```python
svr2=SVR(kernel='poly')
svr2.fit(x_scaled,y_scaled)
print(svr2.predict([[6.6]]))
```

```python
svr3=SVR(kernel='linear')
svr3.fit(x_scaled,y_scaled)
print(svr3.predict([[6.6]]))
```

```python
# svr=SVR(kernel='rbf') radial bases function
plt.scatter(x_scaled,y_scaled,color='red')
plt.plot(x_scaled,svr.predict(x_scaled),color='blue')
plt.show()

# svr=SVR(kernel='poly') non-linear
plt.scatter(x_scaled,y_scaled,color='red')
plt.plot(x_scaled,svr2.predict(x_scaled),color='blue')
plt.show()

# svr=SVR(kernel='linear') linear
plt.scatter(x_scaled,y_scaled,color='red')
plt.plot(x_scaled,svr3.predict(x_scaled),color='blue')
plt.show()
```
